# Линейная регрессия на данных роста и веса 25 тысяч подростков

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as opt
%matplotlib inline
```

```python
data = pd.read_csv('./weight_height.csv', index_col='Index')
data.head()
```

Гистрограммы распредления признаков:

Интереснее будет добавить третий признак: индекс массы тела [BMI](https://en.wikipedia.org/wiki/Body_mass_index).

```python
def make_bmi(height_inch, weight_pound):
    METER_TO_INCH, KILO_TO_POUND = 39.37, 2.20462
    return (weight_pound / KILO_TO_POUND) / (height_inch / METER_TO_INCH) ** 2
```

```python
data['BMI'] = data.apply(lambda row: make_bmi(row['Height'], row['Weight']), axis=1)
```

Попарные зависимости признаков:

```python
sns.pairplot(data)
```

Разобьём вес на 3 категории: 1 - если вес меньше 120 фунтов, 3 - если вес больше 150 фунтов, 2 - в остальных случаях. Построим ящик с усами для зависимости роста от весовой категории:

```python
def get_category(weight):
    if weight < 120:
        return 1
    elif weight <= 150:
        return 2
    return 3
```

```python
data['Weight_category'] = data['Weight'].apply(get_category)
sns.boxplot(data['Weight_category'], data['Height'])
```

# Минимизация квадратичной ошибки

Будем приближать зависимость роста $y$ от веса $x$ прямой линией $y = w_0 + w_1 * x$.
Напишем функцию, которая будет вычислять квадратичную ошибку приближения:
    $$loss(w_0, w_1) = \sum_{i=1}^n {(y_i - (w_0 + w_1 * x_i))}^2 $$ 

```python
def loss(w):
    w0, w1 = w
    return ((data['Height'] - (w0 + w1 * data['Weight'])) ** 2).sum()
```

Посмотрим на зависимость функции от аргументов:

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

X = np.arange(-10, 10)
Y = np.arange(-10, 10)
X, Y = np.meshgrid(X, Y)
Z = np.zeros(X.shape)

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i][j] = loss((X[i][j], Y[i][j]))
        
ax.plot_surface(X, Y, Z)
ax.set_xlabel('Intercpet')
ax.set_ylabel('Slope')
ax.set_zlabel('Loss')
```

Найдем оптимальные значения $w_0$ и $w_1$:

```python
minimum = opt.minimize(loss, [0,0], bounds=((-100,100), (-5,5)), method='L-BFGS-B')
w0_opt, w1_opt = minimum.x
```

Построим прямую на данных:

```python
x = np.linspace(40, 200)
y = w0_opt + w1_opt * x

data.plot(y='Height', x='Weight', kind='scatter', title='Height vs weight', label='Data')
plt.plot(x, y, label='Predict', color='y')
plt.legend()
```
