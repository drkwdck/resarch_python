# Центральная предельная теорема

Проверка работы центральной предельной теоремы

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

## Экспоненциальное распределение

### Гистограмма и теоретическая плотность распределния для выборки

Сгенерируем выборку объема 1000:

```python
exponential_rv = sts.expon()
sample = exponential_rv.rvs(1000)
```

Построим гистограмму и теоритическую полтность распределения:

```python
x = np.linspace(0, 10, 500)
pdf = exponential_rv.pdf(x)

plt.plot(x, pdf, label='theoretical CDF')
plt.hist(sample, density=True, label='fraction of sample')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
```

### Гистограмма распределний выборочных средних для 1000 выборок разных объёмов

Сгенерируем выборку выборочных средних объёма 1000 из выборок объёма n.
Построим гистограмму распределения выборочного среднего  и добавим график полтности вероятности
нормального распределния, к которому приближается распределение выборочных средних, согласно
центральной предельной теореме:

```python
def foo(n):
    sample_mean = np.array([exponential_rv.rvs(n).mean() for i in range(1000)])
    norm_rvs = sts.norm(1,1/np.sqrt(n))
    x = np.linspace(0, 4, 100)
    pdf = norm_rvs.pdf(x)

    plt.hist(sample_mean, density=True, label='fraction of sample mean')
    plt.plot(x, pdf, label='theoretical pdf')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('f(x)')
```

Если n = 5:

```python
foo(5)
```

Если n = 10:

```python
foo(10)
```

Если n = 50:

```python
foo(50)
```

С ростом объёма выборки исходного распределения (n) выборочные средние всё больше подчиняются нормальному распредлению.
$\bar X \sim N(\mu, \sigma^2 / \sqrt n)$
